```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 201)
    
# Visualization
plt.figure(figsize=(4, 8))

plt.plot(x,x,label=r'$x$')
plt.plot(x,x**2,label=r'$x^2$')
plt.plot(x,x**3,label=r'$x^3$')
plt.plot(x,x**4,label=r'$x^4$')
plt.plot(x,x**5,label=r'$x^5$')
plt.plot(x,x**6,label=r'$x^6$')

plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")
plt.ylim(-1,1)
plt.legend(loc='upper right')
#plt.axis('equal')
plt.grid()
plt.show()

```

```python
# Visualization
plt.figure(figsize=(4, 8))
fig, axs = plt.subplots(2)

x = np.linspace(-np.pi, np.pi, 501)


axs[0].plot(x,np.sin(x),label=r'$\sin(x)$')
axs[0].plot(x,np.sin(2*x),label=r'$\sin(2x)$')
axs[0].plot(x,np.sin(3*x),label=r'$\sin(3x)$')
axs[0].plot(x,np.sin(4*x),label=r'$\sin(4x)$')
axs[1].plot(x,np.cos(x),label=r'$\cos(x)$')
axs[1].plot(x,np.cos(2*x),label=r'$\cos(2x)$')
axs[1].plot(x,np.cos(3*x),label=r'$\cos(3x)$')
axs[1].plot(x,np.cos(4*x),label=r'$\cos(4x)$')

axs[0].xlabel("")
axs[0].ylabel(r"$\psi(x)$")
axs[0].legend(loc='upper right')
axs[0].grid()

axs[1].xlabel(r"$x$")
axs[1].ylabel(r"$\psi(x)$")
axs[1].legend(loc='upper right')
axs[1].grid()

plt.show()

```
